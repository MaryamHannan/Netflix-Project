Q. 3) How many Movies & TV Shows are in the dataset ? Show with Bar Graph.

```python
movies_count = (netflix['Category'] == 'Movie').sum()
tv_shows_count = (netflix['Category'] == 'TV Show').sum()
```

```python
# Create a bar graph to visualize the counts
import matplotlib.pyplot as plt
plt.figure(figsize=(5, 5))
plt.bar(['Movies', 'TV Shows'], [movies_count, tv_shows_count], color=['orange', 'blue'])
plt.title('Number of Movies and TV Shows in the Dataset')
plt.xlabel('Category')
plt.ylabel('Count')
plt.show()
```

```python
# other method 
movie=netflix.groupby('Category').Category.count()
```

```python
plt.figure(figsize=(5, 5))
sns.countplot(x='Category', data=netflix)
plt.title('Number of Movies and TV Shows in the Dataset')
```
